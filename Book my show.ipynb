```python
import pandas,requests
from bs4 import BeautifulSoup
```

```python
url="https://in.bookmyshow.com/national-capital-region-ncr/movies/comingsoon"
request=requests.get(url)
```

```python
htmldata=BeautifulSoup(request.text,'html.parser')
```

```python
movie=[]
day=[]
language=[]
for m in htmldata.findAll(class_='card-right'):
    movie.append(m.find('h4').text) 
    day.append(m.find(class_='day').text)
    language.append(m.find(class_='__language').text)
```

```python
data=pandas.DataFrame({'Movie': movie,
                'Day of Release':day,
                 'Language':language})
```

```python
pandas.set_option('display.max_rows', None)
data.to_csv('Movie_Data.csv')
data
```
