```python
import requests
from bs4 import BeautifulSoup
```

```python
dataapi="https://covidtracking.com/api/v1/states/current.json"
nameapi="https://covidtracking.com/api/v1/states/info.json"
```

```python
data=requests.get(dataapi)
namedata=requests.get(nameapi)
```

```python
print(data,namedata)
```

```python
jsondata=data.json()
jsonname=namedata.json()
```

```python
import pandas as pd
```

```python
alldata = pd.DataFrame(jsondata)
statedata = pd.DataFrame(jsonname)
```

```python
alldata = alldata[['state','positive','recovered','death']]
statedata = statedata[['state','name']]
```

```python
finaldata = pd.merge(statedata,alldata,on='state')
```
